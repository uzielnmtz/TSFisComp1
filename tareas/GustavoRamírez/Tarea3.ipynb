```julia
function mapeo(f::Function,x0,iteraciones)
n=[]  
for _ in 1:iteraciones
    xn=f(x0)
    x0=xn
    push!(n,x0)
end

return x0,n
end
```

```julia

"""Esta función nos permite obtener x_n para un número determinado de iteraciones en un mapeo,
además nos permite realizar el análisis gráfico del mapeo para el número dado de iterados de una órbita y poder visualizarlos 
en el eje de las abcisas de manera automática"""

#Inicia la función
function analisis_mapeo2(f::Function,x0,iteraciones::Int)
x01=x0
n=[]
w=[]   
v=[]
    
#Esta parte nos permite capturar el número de iteraciones en el vector n.
for _ in 1:iteraciones
    xn=f(x0)
    x0=xn
    push!(n,x0)

end
   
#Esta parte nos permite capturar cada una de las componentes del vector n[a] que está regido por la sucesión a que va como {1,1,
#2,2,3,3,4,4,...} en el vector v,el cual nos servirá posteriromente para graficarlo en el eje y.
for i in 2:2*iteraciones 
    a=(1/4)*(2*i + (-1)^(i+1) + 1)
    push!(v,n[Int(a)])
            
end

#Esta parte nos permite del mismo modo capturar las componentes del vector n[a] que está regido por la sucesión a que va como 
#{1,1,2,2,3,3,4,...} en el vector v, pero en esta ocasión la sucesión termina un número antes por lo cual es necesario restarle
#1 al límite superior del ciclo for, esto nos servirá posteriormente para graficarlo en el eje x.
for i in 1:(2*iteraciones-1) 
    a=(1/4)*(2*i + (-1)^(i+1) + 1)
    push!(w,n[Int(a)])
            
end
    
plot(0.0:1/32:5.2, f, xaxis=(L"x", 0:5.2), yaxis=(L"F(x)",(0.0,2.0)), legend=false, title=L"F(x)=\sqrt{x}", grid=false)
plot!(0:5, x->x, color=:red) 

plot!([x01,x01,n[1]], [-0.5,n[1],n[1]], line=(:green, :path, 2.0, :dot), marker=(:dot, 0.5))
plot!([w], [v], line=(:green, :path, 2.0, :dot), marker=(:dot, 0.5))


end


```

Para la segunda parte:

Vamos a utilizar el método de Newton, utilizando iterados iniciales seleccionados de manera particular para obtener los puntos fijos.

```julia

struct Dual
    x  :: Float64
    x′ :: Float64
end
```

```julia
dual(x0::Real) = Dual(x0, 1)
```

```julia
+(u::Dual, v::Dual) = Dual( u.x + v.x, u.x′ + v.x′)
+(a::Real, u::Dual) = Dual( a + u.x, u.x′)
+(u::Dual, a::Real) = Dual( a + u.x, u.x′)

-(u::Dual, v::Dual) = Dual( u.x - v.x, u.x′ - v.x′)
-(u::Real, v::Dual) = Dual( u - v.x, - v.x′)
-(u::Dual, v::Real) = Dual( u.x - v, u.x′ )

*(u::Dual, v::Dual) = Dual( u.x * v.x, u.x * v.x′ + u.x′ * v.x)
*(u::Real, v::Dual) = Dual( u * v.x, u * v.x′ )
*(u::Dual, v::Real) = Dual( u.x * v, u.x′ * v)

function /(u::Dual, v::Dual)
    y = u.x / v.x
    Dual( y, (u.x′ - y * v.x′)/v.x )
end

function /(u::Real, v::Dual)
    y = u / v.x
    Dual( y, -(u*v.x′)/(v.x)^2 )
end
function /(u::Dual, v::Real)
    y = u.x / v
    Dual( y, (u.x′)/v )
end
```

No se definió el operador ^ ya que en esta ocasión no será necesario.

```julia
function newton(f::Function,x0,iteraciones::Int)   
   
x=x0
contador=1
    
    while(contador <= iteraciones) 
        x = x0 - f(dual(x0)).x/f(dual(x0)).x′
        
        x0=x
        
     contador = contador + 1
    end
    return x
    
end
```

```julia
f(x) = x^2-x-1.1
newton.(f,[-2,1.2],100)
```

Por lo tanto, utilizando el método de Newton los puntos fijos son:

$x \approx -0.661895$

$x \approx 1.6619$

Los puntos fijos obtenidos de manera analítica coinciden con los puntos fijos obtenidos a partir del método de Newton.

```julia
function mapeo_1(f::Function,x0,iteraciones::Int)
n=[]  

for _ in 1:iteraciones
    xn=f(x0)
    x0=xn
    push!(n,x0)
end

return n
end
```

Analicen los mapeos

- F(x) = x^2,

- F(x) = x^3,

- F(x) = -x^3,

que se satisface $F(0) = F'(0) = 0$. ¿Qué podemos decir de la convergencia?

Para observar la dinámica de estos mapeos, utilizaremos esta vez la función propuesta por el notebook correspondiente a esta tarea.

Para $F(x)=x^2$ :

```julia
f(x)=x^2
```

```julia
"""
    iteramapeo(f, x0, n)

Esta función itera el mapeo `f`, usando como condición inicial `x0`, `n::Int` veces
y regresa dos vectores que continen la secuencia de puntos, incluyendo las
*visitas* a la diagonal.
"""
function iteramapeo(f, x0, n::Int)
    xx = [x0]
    fx = [0.0]
    for it = 1:n
        x1 = f(x0)
        push!(xx, x0)
        push!(fx, x1)
        x0 = x1
        push!(xx, x1)
        push!(fx, x1)
    end
    xx, fx
end
```

```julia
x1, fx1 = iteramapeo(f, 0.8, 20); 
x2, fx2 = iteramapeo(f, -0.9, 20);
```

```julia
plot(-2:1/32:2, f, label=L"F_\alpha(x)",
    xaxis=(L"x", (-2.0, 2.0)), yaxis=(L"f(x)", (-0.5,1.5)))
plot!(-2:1/32:2, identity, label=L"Id")
plot!(x1, fx1, marker=(:dot, 3, 0.4), label=L"x1\alpha")
plot!(x2, fx2, marker=(:dot, 3, 0.4), label=L"x2\alpha")
```

Se puede observar que los iterados de un punto cercano al punto fijo $x=0$ son atraídos por ambos lados, mientras que el punto fijo $x=1$ repele a los iterados.
Es decir, los iterados convergen al punto fijo $x=0$.
Esto también puede observarse de la información que nos proporciona la magnitud de la derivada, dado que $|F'(0)| = 0 < 1$ entonces $x=0$ es un punto fijo atractor, mientras que $|F'(1)|= 2 > 1$ por lo que $x=1$ es un punto fijo repulsor.

Para $F(x) = x^3$ :

```julia
f(x)=x^3
```

```julia
x1α, fx1α = iteramapeo(f, 0.8, 20); 
x2α, fx2α = iteramapeo(f, -0.9, 20);
```

```julia
plot(-2:1/32:2, f, label=L"F_\alpha(x)",
    xaxis=(L"x", (-2.0, 2.0)), yaxis=(L"f(x)", (-1.5,1.5)))
plot!(-2:1/32:2, identity, label=L"Id")
plot!(x1α, fx1α, marker=(:dot, 3, 0.4), label=L"x1\alpha")
plot!(x2α, fx2α, marker=(:dot, 3, 0.4), label=L"x2\alpha")
```

Como se puede apreciar, los puntos cercanos al punto fijo $x=0$ son atraídos; es decir, conforme aumenta el número de iteraciones se aproximan más al punto fijo $x=0$ o dicho de otra manera convergen a $x=0$.
Por otro lado, el punto fijo $x=-1$ los repele de igual modo que el punto fijo $x=1$.

Esto también puede observarse de la información que nos proporciona la magnitud de la derivada, dado que $|F'(0)| = 0 < 1$ entonces $x=0$ es un punto fijo atractor, mientras que $|F'(-1)|= 3 > 1$ por lo que $x=-1$ es un punto fijo repulsor y 
$|F'(1)|= 3 > 1$ por lo que $x=1$ también es un punto fijo repulsor.

Para $F(x) = -x^3$ :

```julia
f(x)=-x^3
```

```julia
x1β, fx1β = iteramapeo(f, 0.8, 20); 
x2β, fx2β = iteramapeo(f, -0.9, 20);
```

```julia
plot(-2:1/32:2, f, label=L"F_\beta(x)",
    xaxis=(L"x", (-2.0, 2.0)), yaxis=(L"f(x)", (-1.5,1.5)))
plot!(-2:1/32:2, identity, label=L"Id")
plot!(x1α, fx1α, marker=(:dot, 3, 0.4), label=L"x1\beta")
plot!(x2α, fx2α, marker=(:dot, 3, 0.4), label=L"x2\beta")
```

Como podemos ver, los puntos cercanos al punto fijo $x=0$ son atraídos; es decir, conforme aumenta el número de iteraciones se aproximan más al único punto fijo $x=0$ o dicho de otra manera convergen a $x=0$.

Esto también puede observarse de la información que nos proporciona la magnitud de la derivada, dado que $|F'(0)| = 0 < 1$ entonces $x=0$ es un punto fijo atractor.

Este caso en particular a pesar de su sencillez es bastante curioso debido a que la función se encuentra sobre la misma recta identidad, por lo que todos los puntos de $F(x)=x$ son puntos fijos neutros; es decir dado que $|F'(x)|=1$  $\forall x$, entonces $\forall x$ $x$ es un punto neutro.

```julia
f(x)=-x
```

```julia
x1δ, fx1δ = iteramapeo(f, 0.7, 20); 
x2δ, fx2δ = iteramapeo(f, -0.9, 20);
x3δ, fx3δ = iteramapeo(f, 0.1, 20);
```

```julia
function estabilidad(f::Function,x0)
    if(abs(f(f(dual(x0))).x′) < 1)
        println(x0 , " es un punto de periodo dos atractor o estable. ")
        println(" ")
        
    elseif(abs(f(f(dual(x0))).x′) > 1)
        println(x0 , " es un punto de periodo dos repulsor o inestable. ")
        println(" ")
    
    end
end
```
